```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas as pd
import lmfit
import corner
```

```python
#Jassal parametrization w(z)=w_0 + w_1{z \over {(1+z)}^2}
def w(z1,w0,w1):
    wz=w0 + w1*(z1/(1+z1)**2)
    return wz
```

```python
#Define  random redshifts
num_zs = 1000
zs= np.linspace(0.01, 10., num=num_zs)
```

```python
w_0=-0.96
w_1=-0.28
```

```python
wz1=w(zs,w_0,w_1)
```

```python
plt.plot(zs,wz1, '-')
```
